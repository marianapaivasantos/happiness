```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path1 = r"C:\Users\ShannonKnight\OneDrive - PPO Serve\Documents\ironhack\Week 4\project_2\data\clean data\avgbyregion_2015.csv"
path2 = r"C:\Users\ShannonKnight\OneDrive - PPO Serve\Documents\ironhack\Week 4\project_2\data\clean data\avgbyregion_2019.csv"

avg2015 = pd.read_csv(path1)
avg2019 = pd.read_csv(path2)
```

```python
avg2015.dtypes
```

```python
avg2015_westeur = avg2015[avg2015['region'] == 'Western Europe']
```

```python
avg2015_africa = avg2015[avg2015['region'] == 'Sub-Saharan Africa']
```

```python
avg2019_westeur = avg2019[avg2019['region'] == 'Western Europe']
```

```python
avg2019_africa = avg2019[avg2019['region'] == 'Sub-Saharan Africa']
```

```python
print(avg2015_africa['happiness_score'].unique())
```

```python
# Custom colors (pastel blue and pastel orange)
pastel_blue = (0.6, 0.8, 1)
pastel_orange = (1, 0.6, 0.4)

# Sort the DataFrame by 'happiness_score' column in descending order
avg2015_africa_sorted = avg2015_africa.sort_values(by='happiness_score', ascending=False)

# Sample data for the year 2015
x_values = avg2015_africa_sorted['country']
y1_values = avg2015_africa_sorted['happiness_score']  # Bar plot data
y2_values = avg2015_africa_sorted['avg_hap_score_2015']    # Line plot data

# Create bar plot with pastel blue color
sns.barplot(x=x_values, y=y1_values, color=pastel_blue)

# Create line plot with pastel orange color
sns.lineplot(x=x_values, y=y2_values, color=pastel_orange, marker='o')

# Set labels and title
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('Happiness Scores in 2015 - Sub Saharan Africa')

# Rotate x-axis labels to avoid overlap
plt.xticks(rotation=90)

# Show plot
plt.show()
```
