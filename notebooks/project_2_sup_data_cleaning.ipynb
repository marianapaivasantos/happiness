```python
#pivot_data = data.pivot(index="Country Name",columns="Series Name",values="2015 [YR2015]" )

# Pivot the data for 2015
pivot_2015 = pd.pivot_table(df[df['2015 [YR2015]'] != '..'], 
                             index=['Country Name'], 
                             columns='Series Name', 
                             values='2015 [YR2015]')

pivot_2015_reset = pivot_2015.reset_index(level='Country Name')

# Pivot the data for 2019
pivot_2019 = pd.pivot_table(df[df['2019 [YR2019]'] != '..'], 
                             index=['Country Name'], 
                             columns='Series Name', 
                             values='2019 [YR2019]')

pivot_2019_reset = pivot_2015.reset_index(level='Country Name')
```

```python
# Check if the 'Country Name' is in the list of countries to keep
mask1 = pivot_2015_reset['Country Name'].isin(countries_to_keep)
mask2 = pivot_2019_reset['Country Name'].isin(countries_to_keep)

# Filter the DataFrame to keep only the rows where the 'Country Name' is in the list of countries to keep
pivot_2015_filtered = pivot_2015_reset[mask1]
pivot_2019_filtered = pivot_2019_reset[mask1]
```
