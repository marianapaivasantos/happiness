```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Scatter plot of happiness scores vs. GDP per capita for 2015
plt.figure(figsize=(10, 6))
plt.scatter(df_2015['gdp_per_capita'], df_2015['happiness_score'], color='blue', label='2015')
plt.title('Happiness Score vs. GDP per Capita (2015)')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Happiness Score')
for i, country in enumerate(df_2015['country']):
    plt.annotate(country, (df_2015['gdp_per_capita'][i], df_2015['happiness_score'][i]))
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Sorting data by GDP per capita for better visualization
df_2015_sorted = df_2015.sort_values(by='gdp_per_capita')

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.barh(df_2015_sorted['country'], df_2015_sorted['happiness_score'], color='blue')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.title('Happiness Scores vs. Countries (2015)')
plt.grid(axis='x')

plt.show()
```

```python
df_sorted_2015 = df_2015.sort_values(by='happiness_score', ascending=False)

# Select the top 5 most happy and least happy countries
top_5_most_happy_2015 = df_sorted_2015.head(5)
top_5_least_happy_2015 = df_sorted_2015.tail(5)

# Scatter plot of unemployment vs. happiness for the top 5 most and least happy countries
plt.figure(figsize=(10, 6))

# Plot top 5 most happy countries in green
plt.scatter(top_5_most_happy_2015['unemployment_pc'], top_5_most_happy_2015['happiness_score'], color='green', label='Top 5 Most Happy')

# Plot top 5 least happy countries in red
plt.scatter(top_5_least_happy_2015['unemployment_pc'], top_5_least_happy_2015['happiness_score'], color='red', label='Top 5 Least Happy')

for x, y, country in zip(top_5_most_happy_2015['unemployment_pc'], top_5_most_happy_2015['happiness_score'], top_5_most_happy_2015['country']):
    plt.annotate(country, (x, y), textcoords="offset points", xytext=(3,3), ha='center')

for x, y, country in zip(top_5_least_happy_2015['unemployment_pc'], top_5_least_happy_2015['happiness_score'], top_5_least_happy_2015['country']):
    plt.annotate(country, (x, y), textcoords="offset points", xytext=(5,5), ha='center')


plt.title('Unemployment vs. Happiness for Top 5 Most and Least Happy Countries - 2015')
plt.xlabel('Unemployment Rate')
plt.ylabel('Happiness Score')
plt.legend()
plt.grid(True)
plt.show()
```
